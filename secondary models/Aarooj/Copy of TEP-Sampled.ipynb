```python
# Define the file path
file_path = 'F:\TEP_Faulty_Training.csv'

# Define the chunk size
chunk_size = 1000  # Approximately 1% of your data

# Create an empty DataFrame to hold chunks
df_chunks = []

# Read the file in chunks
for chunk in pd.read_csv(file_path, chunksize=chunk_size):
    df_chunks.append(chunk)

# Concatenate all chunks into a single DataFrame
full_df_1 = pd.concat(df_chunks, ignore_index=True)

# Now you can work with 'full_df' for further analysis
print(full_df_1.shape)
b2 = full_df_1
```

```python
# Define the file path
file_path = 'F:\TEP_Faulty_Testing.csv'

# Define the chunk size
chunk_size = 1000  # Approximately 1% of your data

# Create an empty DataFrame to hold chunks
df_chunks = []

# Read the file in chunks
for chunk in pd.read_csv(file_path, chunksize=chunk_size):
    df_chunks.append(chunk)

# Concatenate all chunks into a single DataFrame
full_df = pd.concat(df_chunks, ignore_index=True)

# Now you can work with 'full_df' for further analysis
print(full_df.shape)
b4 = full_df
```

```python
b1 = a1['fault_free_training']
b3 = a3['fault_free_testing']
```

```python
frames_tr = [b1,b2]
train_ts = pd.concat(frames_tr)
```

```python
frames_ts = [b3,b4]
test = pd.concat(frames_ts)
```

```python
print('Shape of the Train dataset:', train_ts.shape)
print("Shape of the Test dataset:", test.shape)
```

```python
print("Distrubution of faults in train dataset:")
print(train_ts['faultNumber'].value_counts())
```

```python
print("Distrubution of faults in train dataset:")
print(test['faultNumber'].value_counts())
```

```python
Sampled_train = pd.DataFrame() # dataframe to store the train dataset
Sampled_test = pd.DataFrame() # dataframe to store test
Sampled_cv = pd.DataFrame() # dataframe to store cv data
# Program to construct the sample train data
frame = []
for i in set(train_ts['faultNumber']):
    b_i = pd.DataFrame()
    if i == 0:
        b_i = train_ts[train_ts['faultNumber'] == i][0:20000] # 20000 --- >25*480
#         frame.append(b_i)

    else:
        fr = []
        b = train_ts[train_ts['faultNumber'] == i]
        for x in range(1,26):
            b_x = b[b['simulationRun'] == x][20:500]
            fr.append(b_x)


        b_i = pd.concat(fr)


    frame.append(b_i)
Sampled_train = pd.concat(frame)
# print(Sampled_train['faultNumber'].value_counts())
```

```python
# Program to construct the sample CV Data
frame = []
for i in set(train_ts['faultNumber']):
    b_i = pd.DataFrame()
    if i == 0:
        b_i = train_ts[train_ts['faultNumber'] == i][20000:30000] # 30000 ---> 20000+10*480
#         frame.append(b_i)
    else:
        fr = []
        b = train_ts[train_ts['faultNumber'] == i]
        for x in range(26,36):
            b_x = b[b['simulationRun'] == x][20:500]
            fr.append(b_x)

        b_i = pd.concat(fr)

    frame.append(b_i)
Sampled_cv = pd.concat(frame)
# print(Sampled_cv['faultNumber'].value_counts())
```

```python
# Program to construct the sample test Data
frame = []
for i in set(test['faultNumber']):
    b_i = pd.DataFrame()
    if i == 0:
        b_i = test[test['faultNumber'] == i][0:2000] # 20000 ---> 10*480
#         frame.append(b_i)
    else:
        fr = []
        b = test[test['faultNumber'] == i]
        for x in range(1,11):
            b_x = b[b['simulationRun'] == x][160:640]
            fr.append(b_x)

        b_i = pd.concat(fr)

    frame.append(b_i)
Sampled_test = pd.concat(frame)
# print(Sampled_test['faultNumber'].value_counts())
```
